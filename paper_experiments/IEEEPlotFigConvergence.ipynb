```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
#plt.rcParams['mathtext.fontset'] = 'cm'
#plt.rcParams['font.family'] = 'STIXGeneral'
import scienceplots
plt.style.use('ieee')
```

```python
#font = {'family' : 'STIXGeneral',
#        'weight' : 'bold',
#        'size'   : 12}

#matplotlib.rc('font', **font)

x = np.array([100,500,1000])
```

```python
fedbuff_samples = np.array([[5750, 5100, 4450], [7500, 9200, 6350],[15500, 11500, 18850]]) * 10 / 1000
qafel_samples = np.array([[4650, 8200, 3750], [13900, 6400, 8900],[18650, 14100, 12400]]) * 10 / 1000
```

```python
fedbuff = np.mean(fedbuff_samples, axis=1)
error_fedbuff = np.std(fedbuff_samples, axis=1)
qafel = np.mean(qafel_samples, axis=1)
error_qafel = np.std(qafel_samples, axis=1)
print(max(qafel/fedbuff))
print(min(qafel/fedbuff))
```

```python
plt.plot(x, fedbuff, '-o', label='FedBuff')
plt.fill_between(x, fedbuff-error_fedbuff, fedbuff+error_fedbuff, interpolate=True, alpha=.3)
plt.plot(x, qafel, '--s', label='QAFeL')
plt.fill_between(x, qafel-error_qafel, qafel+error_qafel, interpolate=True, alpha=.3)
plt.ylabel('Client uploads (in thousands)', fontweight='bold')
plt.xlabel('Concurrency', fontweight='bold')
plt.xticks(x)
plt.grid()
plt.legend(loc=2)
plt.savefig("tripsCIFAR.pdf", bbox_inches='tight')
plt.show()
```

```python
message_size = 126376 / 1e6
quantized_message_size = 16564 / 1e6
```

```python
fedbuff_size = fedbuff * message_size
error_fedbuff_size = error_fedbuff * message_size
qafel_size = qafel * quantized_message_size
error_qafel_size = error_qafel * quantized_message_size
print(max(fedbuff_size/qafel_size))
print(min(fedbuff_size/qafel_size))
```

```python
plt.plot(x, fedbuff_size, '-o', label='FedBuff')
plt.fill_between(x, fedbuff_size - error_fedbuff_size, fedbuff_size + error_fedbuff_size, interpolate=True, alpha=.3)
plt.plot(x, qafel_size, '--s', label='QAFeL')
plt.fill_between(x, qafel_size - error_qafel_size, qafel_size + error_qafel_size, interpolate=True, alpha=.3)
plt.ylabel('MB uploaded', fontweight='bold')
plt.xlabel('Concurrency', fontweight='bold')
plt.xticks(x)
plt.grid()
plt.legend(loc=2)
plt.savefig("mbsCIFAR.pdf", bbox_inches='tight')
plt.show()
```
